```python
import yfinance
import matplotlib.pyplot as plt
import datetime
import shutil
```

```python
codigo = input("Digite o ticker da ação: ")

dados = yfinance.Ticker(codigo).history("9mo")
```

```python
fechamento = dados.Close
```

```python
maxima = fechamento.max()
minima = fechamento.min()
atual = fechamento[-1]
```

```python
data_final = datetime.date.today()
data_inicial = data_final - datetime.timedelta(days=9*30)

data = yfinance.download(codigo, start=data_inicial, end=data_final)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Close'])
plt.title(f"Preço de Fechamento de {codigo}")
plt.xlabel("Data")
plt.ylabel("Preço de Fechamento")

plt.savefig(f'{codigo}_chart.png')

plt.show()
```
